```python
def get_token_name_from_dexscreener(contract_address):
    url = f"https://dexscreener.com/api/v1/token/{contract_address}"
    headers = {
        'Content-Type': 'application/json',
    }
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        data = response.json()
        if 'data' in data and 'name' in data['data']:
            return data['data']['name']
        else:
            return 'Unknown'
    else:
        return 'Unknown'
```

```python
# Load transactions from CSV file
with open('transactions.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    all_transactions = [row for row in reader]

# Extract swapped tokens
swapped_tokens = []
token_details = {}
for tx in all_transactions:
    tx_hash = tx['tx_hash']
    tx_receipt = w3.eth.get_transaction_receipt(tx_hash)
    block_number = int(tx['block_number'])
    block = w3.eth.get_block(block_number)
    timestamp = datetime.fromtimestamp(block.timestamp)

    for log in tx_receipt.logs:
        # Check if log is a Swap event
        if log.topics[0].hex() == "0xd78ad95fa46c994b6551d0da85fc275fe613ce37657fb8d5e3d130840159d822":
            # Decode event data using Web3's `decode_abi` function
            event_data = eth_abi.decode(['address', 'address', 'uint256'], bytes.fromhex(log.data.hex()[2:]))
            token_contract = log.address

            # Fetch token details from Etherscan API
            response = requests.get(f'https://api.etherscan.io/api?module=token&action=tokeninfo&contractaddress={token_contract}&apikey=N97IVGU5BIJR7P4WYWT7NUSKD3RXBJT1EP')
            token_info = json.loads(response.text)

            if 'result' in token_info and isinstance(token_info['result'], dict):
                token_name = token_info['result'].get('tokenName', 'Unknown')
                token_symbol = token_info['result'].get('symbol', 'Unknown')
            else:
                token_name = 'Unknown'
                token_symbol = 'Unknown'

            swapped_token = {
                'token_name': token_name,
                'token_symbol': token_symbol,
                'sender': event_data[0],
                'receiver': event_data[1],
                'amount': event_data[2],
                'timestamp': timestamp
            }
            swapped_tokens.append(swapped_token)

            # Update token details dictionary
            if token_symbol not in token_details:
                token_details[token_symbol] = {
                    'token_name': token_name,
                    'last_buy': None,
                    'last_sell': None,
                    'holding_period': None
                }

            if swapped_token['sender'] == target_address:
                token_details[token_symbol]['last_sell'] = swapped_token['timestamp']
            elif swapped_token['receiver'] == target_address:
                token_details[token_symbol]['last_buy'] = swapped_token['timestamp']

            # Calculate holding period
            if token_details[token_symbol]['last_buy'] and token_details[token_symbol]['last_sell']:
                holding_period = token_details[token_symbol]['last_sell'] - token_details[token_symbol]['last_buy']
                token_details[token_symbol]['holding_period'] = holding_period

print("Swapped tokens:")
for token in swapped_tokens:
    print(token)

print("\nToken details:")
for symbol, details in token_details.items():
    print(f"Token symbol: {symbol}")
    print(f"Token name: {details['token_name']}")
    print(f"Last buy: {details['last_buy']}")
    print(f"Last sell: {details['last_sell']}")
    print(f"Holding period: {details['holding_period']}")
    print()
```
